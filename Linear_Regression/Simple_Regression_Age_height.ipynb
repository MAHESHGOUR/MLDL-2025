# Importing libraraies

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

## Loading dataset from the csv file

```python
df = pd.read_csv("D:\SIStec\ML\Regression\datasets\synthetic_age_height_data.csv")
```

```python
df.head(10)
```

### Visualize the data using scatter plot, to see the relationship between the age and Height

```python
plt.scatter(df['Age'],df['Height(ft)'])
plt.xlabel('Age')
plt.ylabel('Height')
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

## split the dataset into training and test sets.

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

## Import Linear regression model from the SKlearn Library

```python
from sklearn.linear_model import  LinearRegression
```

## Define the model

```python
model=LinearRegression()
```

## Train the model

```python
model.fit(X_train,y_train)
```

## See the Values of co-efficient/slope.

```python
model.coef_
```

## See the Values of intercept.

```python
model.intercept_
```

## predict the values for test set

```python
y_pred=model.predict(X_test)
```

```python
# single input value
x_single = np.array([[11.0]])  # shape must be (1, 1)

# predict
prediction = model.predict(x_single)
print("Predicted value:", prediction[0])
```

# visulize the best fit line

```python
plt.scatter(df['Age'],df['Height(ft)'])
plt.plot(X_train,model.predict(X_train),color='red')
plt.xlabel('Age')
plt.ylabel('Height(in ft)')
```

# Evaluation Metrics

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
## 🧪 Example Code (Scikit-learn)


from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# y_true = actual values, y_pred = predicted values
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R² Score: {r2:.2f}")
```
